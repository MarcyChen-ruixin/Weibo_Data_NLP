```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import quote

def get_hot_search_json():
    url = 'https://weibo.com/ajax/side/hotSearch'
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        return None
    return response.json()['data']

def get_comments_from_hot_search(hot_search_url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    response = requests.get(hot_search_url, headers=headers)
    if response.status_code != 200:
        return None
    soup = BeautifulSoup(response.text, 'html.parser')
    comments = []
    for p in soup.find_all('p', {'node-type': 'feed_list_content'}):
        text = p.get_text(separator=' ', strip=True)
        comments.append(text)
    return comments

def main():
    hot_search_data = get_hot_search_json()
    if not hot_search_data:
        print('获取微博热搜榜失败')
        return
    num = 50 
    hot_searches = []
    for i, rs in enumerate(hot_search_data['realtime'][:num], 1):
        title = rs['word']
        link = f"https://s.weibo.com/weibo?q={quote(title)}&Refer=top"
        hot_searches.append((title, link))

    for i, (title, link) in enumerate(hot_searches, 1):
        print(f"{i}. {title}")
        print(f"链接: {link}")


if __name__ == '__main__':
    main()
```
